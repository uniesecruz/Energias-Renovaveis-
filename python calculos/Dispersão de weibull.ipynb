```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
vel=np.linspace(2,8,20)
vel.dtype
```

```python
l = [2,4,6]

k= 2
```

```python
def weib(x,n,a):
  p = (a / n) * (x / n)**(a - 1) * np.exp(-(x / n)**a)
  return p
```

```python
dist= np.zeros((len(vel),len(l)), float)
```

```python
for i in range(len(vel)):
  for j in range(len(l)):
    dist[i,j] = weib(vel[i], l[j], k)
```

```python
for j in range(len(l)):
    plt.plot(dist[0:len(vel),j])
```
